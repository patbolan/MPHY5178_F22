```python
import scipy.io
import matplotlib.pyplot as plt
import numpy as np
import requests
import io

# THis is serializing the file from github
file_url = 'https://github.com/patbolan/MPHY5178_F22/blob/main/HW6_data/brain_img.mat?raw=true'
response = requests.get(file_url)
response.raise_for_status()

# Now actualy load the file with scipy
mat = scipy.io.loadmat(io.BytesIO(response.content))
img = mat['img'] # mat is a dictionary that contains the variables

# Show image
plt.imshow(np.abs(img))
plt.show()
```
